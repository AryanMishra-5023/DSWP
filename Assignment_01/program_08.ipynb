# Maximum of 3 numbers

```python
def maximum(a,b,c):
    if a > b and b > c:
        print("%d is maximum" %a)
    elif b > c and c > a:
        print("%d is maximum" %b)
    else:
        print("%d is maximum" %c)
    return;
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))
maximum(x,y,z)
```
