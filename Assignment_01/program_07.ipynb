# Check number is armstrong or not

```python
def armst(num):
    s = 0
    b = num
    cnt = 0
    while num > 0:
        digit = num % 10
        cnt += 1
        num //= 10
    num = b
    while b > 0:
        dig = b % 10
        s += dig * dig * dig
        b //= 10
    if(num == s):
        print("Armstrong number")
    else:
        print("Not an armstrong number")
    return;
a = int(input("Enter a number: "))
armst(a)
```
