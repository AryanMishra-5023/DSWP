# (3)Print reverse number pattern using a for loop

```python
def pattern(x):
    for i in range(0,x + 1):
        for j in range(x - i,0,-1):
            print(j, end = ' ')
        print("")
    return;
rows = int(input("Enter number of rows")) #number of rows is 6
pattern(rows)
```

# (5)Factorial of a nummber

```python
def fact(num):
    f = 1
    for i in range(1,num + 1):
        f *= i
    return f;
a = int(input("Enter a number: ")) #5
print("Factorial is %d" %fact(a))
```

# (6)Check number is palindrome or not

```python
def palindrome(num):
    b = num
    dig = 0
    while num > 0:
        rem = num % 10
        dig = (dig * 10) + rem
        num //= 10
    if(b == dig):
        print("Palindrome")
    else:
        print("Not a palindrome")
    return;

a = int(input("Enter a number: ")) #1221
palindrome(a)
```

# (1)Check number is armstrong or not

```python
def armst(num):
    s = 0
    b = num
    cnt = 0
    while num > 0:
        digit = num % 10
        cnt += 1
        num //= 10
    num = b
    while b > 0:
        dig = b % 10
        s += dig * dig * dig
        b //= 10
    if(num == s):
        print("Armstrong number")
    else:
        print("Not an armstrong number")
    return;
a = int(input("Enter a number: ")) #153
armst(a)
```

# (8) Find the maximum of three numbers.

```python
def maximum(a,b,c):
    if a > b and a > c:
        print("%d is maximum" %a)
    elif b > c and b > a:
        print("%d is maximum" %b)
    else:
        print("%d is maximum" %c)
    return;
x = int(input("Enter first number: ")) #34
y = int(input("Enter second number: ")) #68
z = int(input("Enter third number: ")) #21
maximum(x,y,z)
```
