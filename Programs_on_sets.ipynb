# (2)Add an element to a set

```python
groceries = {"Milk","Bread","Eggs","Chips","Soft Drinks"}
print("Before insertion: ", groceries)
groceries.add("Chocolates")
print("After insertion: ", groceries)
```

# (3)Remove an element from a set

```python
primes = {2,3,5,7,11,13,17,19,23,31}
print("Before deletion: ", primes)
primes.remove(13)
print("After deletion: ", primes)
```

# (6)Check if an element exists in a set

```python
num = {1,2,3,4,5,6,7,8,9}
s = int(input("Enter a number: "))
for i in num:
    if(i == s):
        print("Element exists")
        break;
else:
    print("Element does not exist")
```

# (9)Find the union of two sets

```python
chem = {"Corrosion","Electrochemistry"}
chapters = {"Water hardness","Analytical methods","Nanochemistry"}
print(chem.union(chapters))
```

# (10)Find the intersection of two sets

```python
topics = {"Lime-soda process","Ion-Exchange process","Dissolved Oxygen","EDTA method"}
common = {"EDTA method","Lime-soda process"}
print(topics.intersection(common))
```

# (11)Find the difference between two sets

```python
integers = {45,67,89,32,15,78}
decimals = {34.5,45,46.78,15,78,65.4}
print(decimals.difference(integers))
```

# (12)Find the symmetric difference between two sets

```python
OS1 = {"Android","IOS","Linux","Blackberry"}
OS2 = {"Microsoft","Linux","MacOS"}
print(OS1.symmetric_difference(OS2))
```

# (13) Check if one set is the subset of another

```python
distros = {"Ubuntu","Debian","Kali linux","GNU-linux"}
set_check = {"Ubuntu","Kali linux"}
print(set_check.issubset(distros))
```

# (14)Check if one set is the superset of another

```python
phy = {"Bohr's model"}
topic = {"Uncertainity principle","De-Broglie Hypothesis","Vector model of atom","Bohr's model"}
print(topic.issuperset(phy))
```

# (15)Check if two sets are disjoint

```python
integers = {45,67,89,32,15,78}
prime = {67,23,17}
print(integers.isdisjoint(prime))
```

# (16)Remove duplicates from a list using a set

# (17)Convert a list to a set and back to a list

# (18)Find common elements in two sets using a list

```python
OS1 = {"Android","IOS","Linux","Blackberry"}
OS2 = {"Microsoft","Linux","MacOS"}
OS_1 = list(OS1)
OS_2 = list(OS2)
print("Common elements are:")
for i in (OS_1 or OS_2):
        print(i)
```
