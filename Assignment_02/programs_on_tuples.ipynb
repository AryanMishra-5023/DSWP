# (1)Store and display student information (Name,Age,Grade)

```python
students = [("Aryan",19,"C"),("Om",20,"A"),("Hardik",17,"A")]
print("Student Information:")
for student in students:
    name,age,grade = student
    print(f"Name: {name}, Age: {age}, Grade: {grade}")
```

# (2)List prices of grocery items and total them.

```python
items = [("Bread",24),("Milk",28),("Rice",60),("Wheat",55),("Eggs",20)]
print(items)
(a,b),(c,d),(e,f),(g,h),(i,j) = items
s = b + d + f + h + j
print("Total Price = %.2f" %s)
```

# (3)Pair items with prices using tuples

```python
lists = [("Milk",25),("Eggs",50),("Bread",20)]
print("List of items with prices: ")
for i in lists:
    item,price = i
    print(f"{item}  {price}")
```

# (5)Sort Employee records by salary

```python
Records = [("John",40000),("Alice",55000),("Raj",30000)]
print("Before sorting: ")
print(Records)
(a,b),(c,d),(e,f) = Records
print("After sorting: ")
sorted_employees = sorted(Records, key = lambda x: x[1])
print(sorted_employees)
```

# (6)Count how many Students scored above 75 marks

```python
marks = (67,88,92,74,76,55)
cnt = 0
for i in marks:
    if i > 75:
        cnt = cnt + 1
print("%d Students scored above 75 marks." %cnt)
```

# (7)Create a tuple of stock prices and find the maximum

# (8)Find the Average Temperature

```python
temperatures = (29.5,30.0,32.2,31.5,28.9)
s = 0
for i in temperatures:
    s = s + i
avg = s / len(temperatures)
print("Average temperature is %.2f" %avg)
```
