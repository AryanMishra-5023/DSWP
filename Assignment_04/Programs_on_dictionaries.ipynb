# (1)Write a Python program to create a dictionary from user input.

```python
n = int(input("Enter number of elements: "))
dictionary = {}
for i in range(n):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dictionary[key] = value
print(dictionary)
```

# (2)Write a Python program to count the frequency of each character in a string using a dictionary

```python
n = input("Enter a string: ")
dictionary = {}
for i in n:
    if i in dictionary:
        dictionary[i] += 1
    else:
        dictionary[i] = 1
print(dictionary)
```

# (5)Write a Python program to merge two dictionaries.

```python
n1 = int(input("Enter number of elements(d1): "))
d1 = {}
for i in range(n1):
    key1 = input("Enter key1: ")
    value1 = input("Enter value1: ")
    d1[key1] = value1
n2 = int(input("Enter number of elements(d2): "))
d2 ={}
for j in range(n2):
    key2 = input("Enter key2: ")
    value2 = input("Enter value2: ")
    d2[key2] = value2
print("d1=",d1)
print("d2=",d2)
print("After update: ")
d1.update(d2)
print(d1)
```

# (4)Write a Python program to print the key with maximum and minimum values from a dictionary

```python
dictionary = {'a':456, 'b':321, 'c':78, 'd':678, 'e':56}
maximum = max(dictionary.keys(), key = (lambda x:dictionary[x]))
minimum = min(dictionary.keys(), key = (lambda x:dictionary[x]))
print("Maximum value: ",dictionary[maximum])
print("Minimum value: ",dictionary[minimum])
```

# (6)Write a Python program to convert two lists into a dictionary.

```python
keys = ['id', 'Name', 'Age']
values = [5023, "Aryan", 20]
result = dict(zip(keys, values))
print(result)
```

# (7)Write a Python program to sort a dictionary by its values in ascending order.

```python
data = {'salary1': 30000, 'salary2': 28000, 'salary3': 43000}
sorted_data = dict(sorted(data.items(), key=lambda item: item[1]))
print(sorted_data)
```

# (8)Write a Python program to delete a specified key from a dictionary.

```python
dictionary = {'a': 1, 'b': 2, 'c': 3}
key = 'b'
if key in dictionary:
    del dictionary[key]
print(dictionary)
```

# (9)Write a Python program to count frequency of each word using a dictionary.

```python
text = "My name is Aryan and Aryan is from Palghar"
words = text.split()
freq = {}
for i in words:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
print(freq)
```

# (10) Write a Python program to swap keys and values in a dictionary.

```python
dictionary = {'one': 1, 'two': 2, 'three': 3}
swap = {value: key for key, value in dictionary.items()}
print(swap)
```
